```python
df = pd.read_csv('Stocksdict.csv')
df = df.tail(954)
df
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
from rake_nltk import Rake
from nltk.tokenize import word_tokenize
from nltk import pos_tag

def rakewords(symbol, sentence):
    try:
        print(symbol)
        thesentence = ""

        r = requests.get('https://cloud.iexapis.com/stable/stock/' + str(symbol) + '/company?token=pk_a9918b80cf894ad3bb23d47b3e853ea6')
        content = json.loads(r.content)
        for tag in content['tags']:
            if (tag != 'Miscellaneous'):
                thesentence += tag + " "

        r = Rake(min_length=1, max_length=2)
        text = word_tokenize(sentence)
        labelled = pos_tag(text)

        finalsentence= ''
        for label in labelled:
            if (label[1] != 'VBZ'):
                finalsentence += label[0] + " "
        r.extract_keywords_from_text(finalsentence)
        ranked = r.get_ranked_phrases()
        ranked = [x for x in ranked if not any(c.isdigit() for c in x)][:3]
        return thesentence + " ".join(ranked)
    except:
        return ""


df['keywords'] = df.apply(lambda row: rakewords(row['symbol'], row['descriptions']), axis=1)
df.to_csv('Finaldatav7.csv')
```
